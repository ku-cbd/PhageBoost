```python
from PhageBoost.main import read_sequence_file, call_genes, calculate_features, read_model_from_file, predict, get_predictions
from PhageBoost import get_predictions
import pandas as pd
import xgboost as xgb
```

```python
# you need to install these to work in the PhageBoost kernel (not in the requirements.txt)
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# parameters...


meta = False # Prodigal gene calling in metagenomic?

# rolling params
period = 20
win_type = 'parzen'
min_periods = 1

# region finding params
threshold = 0.8
length = 10
gaps = 5
neighbouring = 0
alpha = 0.001

## not used currently
min_size_of_contig = 10000
search_region = 5500
inwards = 500
n_jobs = 1
```

```python
# files
model_file = '../PhageBoost/models/model_delta_std_hacked.pickled.silent.gz'
fasta_file = '../example/data/NC_000907.fasta.gz'
```

```python
# read fasta
fasta, meta = read_sequence_file(fasta_file)
```

```python
# get genecalls using a wrapper pyrodigal https://github.com/althonos/pyrodigal 
## running the genecalling script twice might break the kernel.
genecalls = call_genes(fasta, meta, min_size_of_contig)
```

```python
# calculate features from gene calls
df = calculate_features(genecalls)
```

```python
# load model
model, feats, feats_, limit = read_model_from_file(model_file)
# transform data
df = get_predictions.get_deltas(df[feats_])
```

```python
# transform single gene predictions to regions
genecalls, nphages, res = predict(model, genecalls, df, 
                                  feats, period, win_type,
                                  min_periods, limit, threshold, 
                                  length, gaps, neighbouring, alpha)
```

# contributions 
takes ages so we reload the model and add more threads

```python
genecalls.index = genecalls['header']
n_jobs=12
model = get_predictions.load_model(model_file,n_jobs)
dtest = xgb.DMatrix(df[feats], feature_names=feats,nthread=n_jobs)
# this will take some time
shap_values = model.predict(dtest,ntree_limit=limit,pred_contribs=True)
```

```python
bias = shap_values[0,-1]
shap_values = pd.DataFrame(shap_values[:,:-1],index=df.index,columns=df.columns)
```

```python
shap_values.head()
```

```python
rolling_shaps =shap_values.rolling(period,win_type=win_type,center=True,min_periods=min_periods).mean()
rolling_dfs =df.rolling(period,win_type=win_type,center=True,min_periods=min_periods).mean()
```

```python
pal = sns.color_palette('tab10', 5)
pal.as_hex()

genecalls['label'] = [pal[0] if i =='0' else pal[1] for i in genecalls['regions'].values]

fig,(ax2,ax1) = plt.subplots(figsize=(20,5),nrows=2,sharex=True)
rolling_shaps.sum(1).plot.bar(stacked=True,ax=ax1,cmap='tab20',width=1)
sns.despine()
ax1.tick_params(axis='x',which='both', bottom=False, top=False, labelbottom=False)
ax2.scatter(y='preds20',x='header',c='label',s=10,data=genecalls)
ax2.tick_params(axis='x',which='both', bottom=False, top=False, labelbottom=False)
plt.xlabel('genecalls')
ax2.set_ylabel('probability')
ax1.set_ylabel('summed contribution');
```
